```python
from datetime import datetime
```

```python

def log_error(function_name, error_type, message):
    with open("error_log.txt", "a", encoding="utf-8") as log:
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        log.write(f'[{timestamp}] Error in {function_name}: {error_type} - {message}\n ')
```

```python
import requests
```

```python
Base_URL = "https://dummyjson.com/products"
```

*Step 1*:  Fetch and Display Products

```python
def fetch_product():
    url = f"{Base_URL}?limit=20"
    try:
        response = requests.get(url, timeout=5)
        data = response.json()
        print("ID |                  TITLE                             |  CATEGORY  |   PRICE(₹)  | RATING")
        for product in data["products"]:
            print(f'{product["id"]:2} | {product["title"]:50} | {product["category"]:10} | {product["price"]:10} | {product["rating"]}')
       
    except requests.exceptions.ConnectionError:
        print("Connection Failed.")
        log_error("fetch_product", "ConnectionError", "No Internet")
    except requests.exceptions.Timeout:
        print("Request Timed out.")
        log_error("fetch_product", "Timeout", "Slow Server")
    except Exception as e:
        print(f"Error: {e}")
        log_error("fetch_product", "Exception", str(e))


fetch_product()
```

*Step 2*: Filter and Sort

```python
def fetch_product():
    url = f"{Base_URL}?limit=20"
    response = requests.get(url, timeout=5)
    data = response.json()
    return data["products"]
    
def filterandSort(products):
    filtered_product = [product for product in products if product["rating"] >= 4.5]
    sorted_product = sorted(filtered_product, key=lambda x: x["price"], reverse=True)
    print("                  TITLE                            |   PRICE(₹) | RATING")
    for product in sorted_product:
        print(f'{product["title"]:50} | {product["price"]:10} | {product["rating"]}')

products = fetch_product()
filterandSort(products)
```

*Step 3*: Search by Category

```python
def fetch_laptop():
    url = f"{Base_URL}/category/laptops"
    try:
        response = requests.get(url, timeout=5)
        data = response.json()
        print("                  TITLE                             |    PRICE(₹) ")
        for product in data["products"]:
            print(f' {product["title"]:50} | {product["price"]:10}')

    except requests.exceptions.ConnectionError:
        print("Connection Failed.")
        log_error("fetch_product", "ConnectionError", "No Internet")
    except requests.exceptions.Timeout:
        print("Request Timed out.")
        log_error("fetch_product", "Timeout", "Slow Server")
    except Exception as e:
        print(f"Error: {e}")
        log_error("fetch_product", "Exception", str(e))  

fetch_laptop()
```

*Step 4*: POST Request 1. & 2.

```python
def add_product():
    url = f"{Base_URL}/add"
    product_to_add = {
  "title": "My Custom Product",
  "price": 999,
  "category": "electronics",
  "description": "A product I created via API"
}

    try:
        response = requests.post(url,json=product_to_add,  timeout=5)
        print("===RESPONSE===")
        print(response.json())

    except requests.exceptions.ConnectionError:
        print("Connection Failed.")
        log_error("fetch_product", "ConnectionError", "No Internet")
    except requests.exceptions.Timeout:
        print("Request Timed out.")
        log_error("fetch_product", "Timeout", "Slow Server")
    except Exception as e:
        print(f"Error: {e}")
        log_error("fetch_product", "Exception", str(e))  

add_product()
```

*Part A*: Guarded Calculator

```python
def safe_divide(a,b):
    try:
        return a / b
    except ZeroDivisionError:
        return("Error: Cannot divide by zero")
    except TypeError:
        return("Error: Invalid input types")

print(safe_divide(10,2))
print(safe_divide(10,0))
print(safe_divide("ten",2))
```

*Part B*: Guarded File Reader

```python
def read_file_safe(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")
  
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))
```

*Part D*: Input Validation Loop:

```python
def product_lookup():
    while True:
        user_input = input("Enter a product ID to look up (1–100), or 'quit' to exit: ")

        if user_input.lower() == "quit":
            break

        if not user_input.isdigit():
            print("Invalid input. Please enter a number")
            continue

        product_id = int(user_input)

        if not 1 <= product_id <= 100 :
            print("ID must be between 1 and 100")
            continue

        try:
            url = f"{Base_URL}/{product_id}"
            response = requests.get(url, timeout=5)
            if response.status_code == 404:
                print("Product not found")
                log_error("lookup_product", "HTTPError", f"404 for ID {product_id}")
            else:
                data = response.json()
                print(f"{data['title']} - ₹{data['price']}")
        
        except requests.exceptions.ConnectionError:
            print("Connection Failed.")
            log_error("lookup_product", "ConnectionError", "No Internet")
        except requests.exceptions.Timeout:
            print("Request Timed out.")
            log_error("lookup_product", "Timeout", "Slow response")
        except Exception as e:
            print(f"Error: {e}")
            log_error("lookup_product", "Exception", str(e))  

product_lookup()
```
